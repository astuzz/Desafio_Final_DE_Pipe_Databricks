```python
from pyspark.sql.functions import col, to_date

df_bronze = spark.table("bronze_legislatura")

df_silver = df_bronze.select(
 "id"
,"dataInicio"
,"dataFim"
).distinct()

# .withColumn para converter as colunas (DataHora) de texto para Timestamp
df_silver = df_silver.withColumn("dataInicio", to_date(col("dataInicio")))
df_silver = df_silver.withColumn("dataFim", to_date(col("dataFim")))

df_silver.write \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .saveAsTable("silver_legislatura")
 
```
