```python
from pyspark.sql.functions import col, to_date, to_timestamp

df_bronze = spark.table("bronze_votacoes")

df_silver = df_bronze.select(
 "id"
,"aprovacao"
,"data"
,"dataHoraRegistro"
,"descricao"
,"proposicaoObjeto"
,"siglaOrgao"
).distinct()

# .withColumn para converter as colunas (DataHora) de texto para Timestamp
df_silver = df_silver.withColumn("data", to_date(col("data")))
df_silver = df_silver.withColumn("dataHoraRegistro", to_timestamp(col("dataHoraRegistro")))

df_silver.write \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .saveAsTable("silver_votacoes")

df_silver.write.mode("overwrite").saveAsTable("silver_votacoes")
```
