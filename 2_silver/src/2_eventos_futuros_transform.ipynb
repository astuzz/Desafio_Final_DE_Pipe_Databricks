```python
# ----------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 2_eventos_futuros_transform
# LOCAL: 2_silver/src/
# OBJETIVO: Converter os tipos e estrutura para ser usada na view pública com calendário de eventos futuros já agendados
# ----------------------------------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col, to_timestamp

df_bronze = spark.table("bronze_eventos_futuros")

df_silver = df_bronze.select(
    to_timestamp(col("dataHoraInicio")).alias("data_hora"),
    col("descricao").alias("pauta"),
    col("descricaoTipo").alias("tipo_evento"),
    col("orgaos").getItem(0).getField("sigla").alias("orgao_sigla"),
    col("situacao")
)

df_silver.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("silver_eventos_futuros")

display(spark.table("silver_eventos_futuros"))
```
