```python


df_bronze = spark.table("bronze_deputados")

df_silver = df_bronze.select(
     "id"
    ,"nome"
    ,"siglaPartido"
    ,"siglaUF"
    ,"idLegislatura"
).distinct()

df_silver.write.mode("overwrite").saveAsTable("silver_deputados")

```
