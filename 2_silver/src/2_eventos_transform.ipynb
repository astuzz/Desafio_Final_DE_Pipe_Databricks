```python

from pyspark.sql.functions import col, to_timestamp

df_bronze = spark.table("bronze_eventos")

 
df_silver = df_bronze.withColumn("id_orgao_organizador", col("orgaos").getItem(0).getField("id"))

 
df_silver = df_silver.select(
    "id",
    "dataHoraInicio",
    "dataHoraFim",
    "descricao",
    "descricaoTipo",
    "situacao",
    "id_orgao_organizador" 
).distinct()

# Conversão de datas
df_silver = df_silver.withColumn("dataHoraInicio", to_timestamp(col("dataHoraInicio")))
df_silver = df_silver.withColumn("dataHoraFim", to_timestamp(col("dataHoraFim")))

 
df_silver.write \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .saveAsTable("silver_eventos")

display(spark.table("silver_eventos"))
```
