```python
from pyspark.sql.functions import col, to_timestamp

df_bronze = spark.table("bronze_proposicoes")

df_silver = df_bronze.select(
 "id"
,"numero"
,"siglaTipo"	 
,"codTipo"
,"dataApresentacao"
,"ano"
,"ementa"
).distinct()

# .withColumn para converter as colunas (DataHora) de texto para Timestamp
df_silver = df_silver.withColumn("dataApresentacao", to_timestamp(col("dataApresentacao")))
  
df_silver.write \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .saveAsTable("silver_proposicoes")
 
```
