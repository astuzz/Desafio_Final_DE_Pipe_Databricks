```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.1b_cpis_membros_silver
# LOCAL: 2_silver/src/5_auditoria_cpis/
# OBJETIVO: Limpeza e padronização dos membros das CPIs
# ENTREGA: Tabela específica de CPIs com membros por fase
# ---------------------------------------------------------------------------------------

from pyspark.sql.functions import col, lit

# Ler a Bronze de Membros (Mock)
df_membros_bronze = spark.table("bronze_cpis_membros")

# Ajustado para os nomes de colunas do Mock
df_membros_silver = df_membros_bronze.select(
    col("id_cpi_proposicao").alias("id_cpi"),
    col("id").alias("id_deputado"),
    col("nome").alias("nome_parlamentar"),
    col("siglaPartido").alias("partido"),
    col("siglaUf").alias("uf"),    
    lit("Membro Proponente").alias("cargo_na_cpi") 
)

# Salvar na Silver
df_membros_silver.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("silver_cpis_membros")

print("Tabela silver_cpis_membros criada com sucesso!")
display(df_membros_silver.limit(10))
```
