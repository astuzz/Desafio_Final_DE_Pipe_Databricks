```python
# -----------------------------------------------------------------------------------------------------------
# SCRIPT: 1.1_frentes_transform
# LOCAL: 2_silver/src/1_atlas_frentes/
# OBJETIVO: Converter os tipos e estrutura referente os dados da tabela Bronze_Frentes.
#           A tabela silver_frentes é utilizada no /3_gold/1_atlas_frentes/1.1_atlas_frentes_consolidado
# ENTREGA: Tabela gold de frentes com membros, partido, UF e legislatura por frente
# -----------------------------------------------------------------------------------------------------------

# Leitura da tabela bronze
df_bronze = spark.table("bronze_frentes")

# Limpeza de dados para a tabela Silver
df_silver = df_bronze.select(
"id"
,"idLegislatura"	
,"titulo"
).distinct()

# Geração da tabela Silver
df_silver.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("silver_frentes")
```
