```python
# --------------------------------------------------------------------------------------------------------------------------------
# SCRIPT: deputados_transform
# LOCAL: 1_silver/src/1_atlas_frentes/
# OBJETIVO: Transformando Dados brutos deputados contidos na tabela gerada pelo deputados_ingest
# --------------------------------------------------------------------------------------------------------------------------------

# Ler dados da bronze
df_bronze = spark.table("bronze_deputados")

# Transformar dados
df_silver = df_bronze.select(
     "id"
    ,"nome"
    ,"siglaPartido"
    ,"siglaUF"
    ,"idLegislatura"
).distinct()

# Gravar dados
df_silver.write.mode("overwrite").saveAsTable("silver_deputados")

```
