```python
# --------------------------------------------------------------------------------------------------
# SCRIPT: 2_membros_frentes_transform
# LOCAL: 2_silver/src/
# OBJETIVO: Converter os tipos e estrutura referente os dados da tabela Bronze_Membros_Frentes
# --------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col

df_bronze = spark.table("bronze_membros_frentes")

# Usamos o alias para deixar os nomes das colunas limpos na Silver
df_silver = df_bronze.select(
    col("id").alias("id_frente"),
    col("titulo").alias("titulo_frente"),
    col("`deputado_.id`").alias("id_deputado"),
    col("`deputado_.nome`").alias("nome_deputado"),
    col("`deputado_.siglaUf`").alias("uf"),
    col("`deputado_.idLegislatura`").alias("id_legislatura"),
    col("`deputado_.codTitulo`").alias("cod_titulo_membro"),
    col("`deputado_.titulo`").alias("titulo_membro")
).distinct()

df_silver.write \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .saveAsTable("silver_membros_frentes")

display(df_silver)
```
