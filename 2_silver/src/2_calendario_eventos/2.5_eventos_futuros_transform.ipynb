```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 2.5_eventos_futuros_transform
# LOCAL: 2_silver/src/2_calendario_eventos/
# OBJETIVO: Converter tipos/estrutura para ser usada na view pública a partir da tabela gerada pelo script 1_bronze/src/2_calendario_eventos/2.5_eventos_futuros_ingest
# ENTREGA: View pública com calendário de eventos futuros já agendados
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col, to_timestamp

# Lê a tabela bronze
df_bronze = spark.table("bronze_eventos_futuros")

# Cria a tabela silver
df_silver = df_bronze.select(
    to_timestamp(col("dataHoraInicio")).alias("data_hora"),
    col("descricao").alias("pauta"),
    col("descricaoTipo").alias("tipo_evento"),
    col("orgaos").getItem(0).getField("sigla").alias("orgao_sigla"),
    col("situacao")
)

# Salva a tabela silver
df_silver.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("silver_eventos_futuros")

display(spark.table("silver_eventos_futuros"))
```
