```python
# ----------------------------------------------------------------------------------------
# SCRIPT: 2.1_orgaos_transform
# LOCAL: 2_silver/src/2_calendario_eventos
# OBJETIVO: Transformar os dados da tabela bronze_orgaos
# ENTREGA: Tabela gold de eventos com dim_orgao, dim_tipo_evento, dim_data
# ----------------------------------------------------------------------------------------

# Carrega a tabela bronze
df_bronze = spark.table("bronze_orgaos")

# Seleciona os campos necessários
df_silver = df_bronze.select(
 "id"
,"codTipoOrgao"
,"nome"
,"nomePublicacao"
,"apelido"
,"sigla"
,"tipoOrgao"
).distinct()

# Salva a tabela silver
df_silver.write.mode("overwrite").saveAsTable("silver_orgaos")
```
