```python
# ----------------------------------------------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 2.1b_dim_tipoevento
# LOCAL: 3_gold/src/2_calendario_eventos
# OBJETIVO: Dados obtidos da tabela silver_referencia_tipoevento gerados pelo script 2_silver/src/2_calendario_eventos/2.1_referencia_tipoevento_transforma
# ENTREGA: Tabela gold de eventos com dim_orgao, dim_tipo_evento, dim_data
# ---------------------------------------------------------------------------------------------------------------------------------------------------------------- 

from pyspark.sql.functions import col

# Ler a tabela da Silver
df_silver = spark.table("silver_referencia_tipoevento")

# Colunas renomeadas
dim_orgao = df_silver.select(
    col("cod").alias("cod_evento"),
    col("nome").alias("evento")
)

# Salvar a tabela Gold
dim_orgao.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gold_dim_tipoevento")

display(spark.table("gold_dim_tipoevento"))
```
