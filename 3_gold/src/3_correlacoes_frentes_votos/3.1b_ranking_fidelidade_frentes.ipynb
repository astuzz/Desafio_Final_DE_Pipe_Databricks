```python
# ----------------------------------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 3.1b_ranking_fidelidade_frentes
# LOCAL: 3_gold/src/3_correlacoes_frentes_votos/
# OBJETIVO: Retorna ranking de fidelidade partidária por frente. Fonte gold_correlacao_frente_voto gerada pelo script 3.1a_alinhamento_frente_partido
# ENTREGA: Verificar se deputados de uma mesma frente votam de forma mais alinhada do que seus colegas de partido.
# ----------------------------------------------------------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col, avg, count, round

# Carrega a tabela de correlação
df_correlacao = spark.table("gold_correlacao_frente_voto")

# Agrupar por frente para calcular a média
df_ranking_fidelidade = df_correlacao.groupBy("nome_frente") \
    .agg(
        count("id_deputado").alias("total_votos_registrados"),
        round(avg("seguiu_partido") * 100, 2).alias("percentual_fidelidade_partidaria")
    ) \
    .orderBy("percentual_fidelidade_partidaria")

# Salva o resultado
df_ranking_fidelidade.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gold_ranking_fidelidade_frentes")

display(df_ranking_fidelidade)
```
