```python
%sql
-- SCRIPT: 6.3_padrao_ausencia_temas
-- LOCAL: 3_gold/src/6_monitor_presenca/
-- OBJETIVO: Obter o percentual de ausência por tema. Usa como fonte de dados a gold_monitor_absenteismo gerada pelo script (3_gold/src/6_monitor_presenca/6.1_analise_absenteismo) e bronze_temas_votacoes gerada pelo script (1_bronze/src/6_monitor_presenca/6.3_ingestao_temas_votacoes)
-- ENTREGA: Detecção de padrão de ausência: deputados que faltam mais em votações específicas

CREATE OR REPLACE TABLE gold_padrao_ausencia_temas AS
SELECT 
    a.nome_deputado,
    a.partido,
    t.tema,
    COUNT(*) as total_votacoes_tema,
    SUM(CASE WHEN a.status_presenca = 'AUSENTE' THEN 1 ELSE 0 END) as total_faltas_no_tema,
    ROUND((SUM(CASE WHEN a.status_presenca = 'AUSENTE' THEN 1 ELSE 0 END) / COUNT(*)) * 100, 2) as percentual_falta_tema
FROM gold_monitor_absenteismo a
JOIN bronze_temas_votacoes t ON a.id_votacao = t.id_votacao
GROUP BY 1, 2, 3
HAVING total_faltas_no_tema > 0
ORDER BY percentual_falta_tema DESC;

SELECT * FROM gold_padrao_ausencia_temas;
```
