```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.4_rede_convocados_cpis
# LOCAL: 3_gold/src/5_auditoria_cpis/
# OBJETIVO: Mapear convocados e suas relações com o setor privado
# ENTREGA: Rede de convocados cruzada com entidades privadas
# ---------------------------------------------------------------------------------------

# Mock de convocados e entidades 
convocados_data = [
    (537330, "Sergio Rial", "Ex-CEO Americanas", "Lojas Americanas S/A"),
    (537357, "João Pedro Stédile", "Líder de Movimento", "MST"),
    (537449, "Glaidson dos Santos", "Sócio-Diretor", "GAS Consultoria (Pirâmide)")
]

df_convocados = spark.createDataFrame(convocados_data, ["id_cpi", "nome_convocado", "vinculo", "entidade_privada"])

# Join com a lista de CPIs para contexto
df_rede_final = spark.table("silver_cpis_detalhes").join(df_convocados, on="id_cpi", how="inner")

# Salvar
df_rede_final.write.mode("overwrite")\
    .option("overwriteSchema", "true")\
    .saveAsTable("gold_cpis_rede_convocados")

print("Rede de convocados e entidades privadas.")
display(df_rede_final)
```
