```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.5_produtividade_cpis
# LOCAL: 3_gold/src/5_auditoria_cpis/
# OBJETIVO: Comparar CPIs com desfecho conclusivo vs não conclusivo
# ENTREGA: Comparativo de produtividade (Relatórios vs Encerradas sem conclusão)
# ---------------------------------------------------------------------------------------

from pyspark.sql.functions import col, when, count

df_gold = spark.table("gold_auditoria_cpis")

# Agregação de produtividade
df_produtividade = df_gold.select("id_cpi", "sigla", "status_cpi").distinct() \
    .withColumn("produtividade", 
                when(col("status_cpi").contains("Relatório"), "Alta (Gerou Relatório)")
                .otherwise("Baixa (Encerrada sem Relatório Final)"))

# Salvar
df_produtividade.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold_cpis_comparativo_produtividade")

print("Comparativo de produtividade final")
display(df_produtividade)
```
