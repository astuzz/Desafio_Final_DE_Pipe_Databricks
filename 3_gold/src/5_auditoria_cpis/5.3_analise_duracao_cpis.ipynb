```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.3_analise_duracao_cpis
# LOCAL: 3_gold/src/5_auditoria_cpis/
# OBJETIVO: Gerar uma base consolidada com todos os CPIs e seus prazos
# ENTREGA: Análise de duração: CPIs que excederam prazo regimental
# ---------------------------------------------------------------------------------------

from pyspark.sql.functions import col

# Ler a base consolidada
df_gold = spark.table("gold_auditoria_cpis")

# Filtrar e formatar apenas a visão de prazos
df_analise_duracao = df_gold.select(
    "id_cpi", "sigla", "descricao", "data_inicio", "data_fim", "duracao_atual_dias", "status_prazo"
).distinct().filter("status_prazo = 'Excedeu Prazo Regimental'")

# Salvar
df_analise_duracao.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold_auditoria_duracao")

print("Entrega 5.3 gerada: Foco em CPIs que excederam o prazo.")
display(df_analise_duracao)
```
