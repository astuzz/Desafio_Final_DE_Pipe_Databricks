```python
# -----------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 5.2_legislacao_derivada_cpis
# LOCAL: 3_gold/src/5_auditoria_cpis/
# OBJETIVO: Cruzar CPIs com proposições legislativas derivadas a partir da tabela gerada na 
#           3_gold\..\5.1_visao_consolidade_cpis
# ENTREGA: Join com proposicoes para identificar legislação derivada de cada CPI
# -----------------------------------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col

# Carregar a visão consolidada da Gold
df_gold_cpis = spark.table("gold_auditoria_cpis")

# Criar dados de Legislação Derivada (Mock)
legis_data = [
    (537357, "PL 1234/2023", "Altera critérios de reintegração de posse rural", "Aguardando Sanção"),
    (537311, "PL 5678/2023", "Tipifica crimes de manipulação de resultados esportivos", "Em Tramitação"),
    (537449, "PL 9012/2023", "Regulamenta o funcionamento de exchanges de criptoativos", "Aprovado")
]

df_legis = spark.createDataFrame(legis_data, ["id_cpi", "projeto_lei", "ementa_lei", "status_lei"])

# Join para identificar qual legislação veio de qual CPI
df_gold_legislacao = df_gold_cpis.join(df_legis, on="id_cpi", how="left")

# Salvar tabela
df_gold_legislacao.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold_cpis_legislacao_derivada")

print("Tabela gold_cpis_legislacao_derivada gerada com sucesso!")
display(df_gold_legislacao.filter("projeto_lei IS NOT NULL"))
```
