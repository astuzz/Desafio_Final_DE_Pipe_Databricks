```python
# --------------------------------------------------------------------------------------------------
# SCRIPT: 1.3_ranking_participacao_frentes
# LOCAL: 3_gold/src/1_atlas_frentes/
# OBJETIVO: Gerar ranking de parlamentares por número de frentes. 
#           (fonte 'gold_atlas_frentes_parlamentares' gerado pelo script 3_gold/1_atlas_frentes/1.1_atlas_frentes_consolidado)
# ENTREGA: Deputados que participam de mais frentes — perfil e temas de interesse
# --------------------------------------------------------------------------------------------------

from pyspark.sql.functions import col, count

df_atlas = spark.table("gold_atlas_frentes_parlamentares")

# Ranking por Deputado
df_ranking = df_atlas\
        .groupBy("id_deputado","nome_deputado","uf") \
        .agg(count("id_frente").alias("total_frentes")) \
        .orderBy(col("total_frentes").desc())

# Salvar dados
df_ranking.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gold_atlas_frentes_parlamentares_ranking")

display(df_ranking)
```
