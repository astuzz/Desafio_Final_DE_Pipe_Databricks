```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.1_visao_consolidada_cpis
# LOCAL: 3_gold/src/5_auditoria_cpis/
# OBJETIVO: Unificar timeline e membros para a visão final de auditoria. Usa tabelas originadas dos scripts
#          /2_silver/src/5_auditoria_cpis/5.1a_cpis_timeline_silver e 5.1b_cpis_membros_silver"
# ENTREGA: Tabela específica de CPIs com timeline e membros por fase
# ---------------------------------------------------------------------------------------

from pyspark.sql.functions import col

# Ler tabelas Silver
df_timeline = spark.table("silver_cpis_detalhes")
df_membros = spark.table("silver_cpis_membros")

# Inner Join para garantir a lista de membros de CPIs mapeadas
df_gold_cpis = df_timeline.join(
    df_membros,
    on="id_cpi",
    how="inner"
).select(
    df_timeline["id_cpi"],
    "sigla",
    "descricao",
    "data_inicio",
    "data_fim",
    "duracao_atual_dias",
    "status_prazo",
    "status_cpi",
    "nome_parlamentar",
    "partido",
    "uf",
    "cargo_na_cpi"
)

# Salvar na Gold
df_gold_cpis.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold_auditoria_cpis")

print("Tabela gold_auditoria_cpis gerada com sucesso!")
display(df_gold_cpis)
```
