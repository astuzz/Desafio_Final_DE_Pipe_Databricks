```python
%sql
-- SCRIPT: 6.2_score_engajamento_parlamentar
-- LOCAL: 3_gold/src/6_monitor_presenca/
-- OBJETIVO: Calcular o score de engajamento tendo como fonte a tabela gerada no script 3_gold/../6.1_analise_absenteismo
-- ENTREGA: Score de engajamento composto: presença × votações × discursos × requerimentos


CREATE OR REPLACE TABLE gold_score_engajamento AS
SELECT 
    id_deputado,
    nome_deputado,
    partido,
    uf,
    COUNT(id_votacao) as total_sessoes,
    SUM(CASE WHEN status_presenca = 'PRESENTE' THEN 1 ELSE 0 END) as presencas_reais,
    -- Cálculo do Score: (Presenças / Total de Sessões) * 10
    ROUND((SUM(CASE WHEN status_presenca = 'PRESENTE' THEN 1 ELSE 0 END) / COUNT(id_votacao)) * 10, 2) as score_engajamento
FROM gold_monitor_absenteismo
GROUP BY 1, 2, 3, 4
ORDER BY score_engajamento DESC;

-- Ranking final
SELECT * FROM gold_score_engajamento;
```
