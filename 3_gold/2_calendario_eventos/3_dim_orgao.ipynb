```python
from pyspark.sql.functions import col

# Ler a tabela da Silver
df_silver = spark.table("silver_orgaos")

# Colunas renomeadas
dim_orgao = df_silver.select(
    col("id").alias("id_orgao"),
    col("nome").alias("nome_orgao"),
    col("sigla").alias("sigla_orgao"),
    col("tipoOrgao").alias("tipo_orgao"),
    col("apelido").alias("apelido_orgao"),
    col("nomePublicacao").alias("nome_publicacao")
)

# Salvar a tabela Gold
dim_orgao.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gold_dim_orgao")

display(spark.table("gold_dim_orgao"))
```
