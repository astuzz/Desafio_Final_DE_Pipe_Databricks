```python
# ----------------------------------------------------------------------------------------
# SCRIPT: 1.1_membros_frentes_ingest
# LOCAL: 1_bronze/src/1_atlas_frentes/
# OBJETIVO: Dados brutos referente aos membros das frentes parlamentares
# ----------------------------------------------------------------------------------------

import pandas as pd

url_csv = "http://dadosabertos.camara.leg.br/arquivos/frentesDeputados/csv/frentesDeputados.csv"

try:
  
    # Separador
    pdf_membros = pd.read_csv(url_csv, sep=';', encoding='utf-8')

    # Converter para Spark DataFrame
    df_membros_completo = spark.createDataFrame(pdf_membros)

    # Salvar na Bronze
    df_membros_completo.write \
        .mode("overwrite") \
        .option("overwriteSchema", "true") \
        .saveAsTable("bronze_membros_frentes")

    print(f"Sucesso! Foram carregados {df_membros_completo.count()} registros via CSV.")

except Exception as e:
    print(f"Erro ao baixar o CSV: {e}")


display(spark.table("bronze_membros_frentes"))
```
