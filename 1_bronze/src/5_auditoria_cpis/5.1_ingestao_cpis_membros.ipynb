```python
# ---------------------------------------------------------------------------------------
# SCRIPT: 5.1_ingestao_cpis_membros (Contingência com MOCK)
# LOCAL: 1_bronze/src/5_auditoria_cpis/
# OBJETIVO: Criar massa de dados para Auditoria de CPIs (Independente de API)
# ENTREGA: Tabela específica de CPIs com timeline de eventos e membros por fase
# ---------------------------------------------------------------------------------------


from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Dados de Lista de CPIs (MOCK)
cpis_data = [
    (537357, "RIC 1/2023", 2023, "CPI - MST", "2023-05-17", "2023-09-27", "Encerrada com Relatório"),
    (537330, "RIC 2/2023", 2023, "CPI - Americanas", "2023-05-17", None, "Em Andamento"),
    (537311, "RIC 3/2023", 2023, "CPI - Manipulação de Resultados", "2023-05-17", "2023-09-26", "Encerrada"),
    (537449, "RIC 4/2023", 2023, "CPI - Pirâmides Financeiras", "2023-06-01", "2023-10-09", "Encerrada")
]

schema_cpis = StructType([
    StructField("id", IntegerType(), True),
    StructField("sigla", StringType(), True),
    StructField("ano", IntegerType(), True),
    StructField("ementa", StringType(), True),
    StructField("dataApresentacao", StringType(), True),
    StructField("dataFim", StringType(), True),
    StructField("status_cpi", StringType(), True)
])

# Dados de Membros (Autores/Membros)
membros_data = [
    (1, "Deputado Ricardo Salles", "PL", "SP", 537357),
    (2, "Deputado Kim Kataguiri", "UNIÃO", "SP", 537357),
    (3, "Deputada Fernanda Melchionna", "PSOL", "RS", 537330),
    (4, "Deputado Gustinho Ribeiro", "REPUBLIC", "SE", 537330),
    (5, "Deputado Felipe Carreras", "PSB", "PE", 537311)
]

schema_membros = StructType([
    StructField("id", IntegerType(), True),
    StructField("nome", StringType(), True),
    StructField("siglaPartido", StringType(), True),
    StructField("siglaUf", StringType(), True),
    StructField("id_cpi_proposicao", IntegerType(), True)
])

# Salvar Tabelas
df_cpis = spark.createDataFrame(cpis_data, schema=schema_cpis)
df_cpis.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze_cpis_lista")

df_membros = spark.createDataFrame(membros_data, schema=schema_membros)
df_membros.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze_cpis_membros")

print("\n[SUCESSO] Tabelas bronze_cpis_lista e bronze_cpis_membros criadas!")
print("O pipeline de Auditoria está DESTRAVADO.")
```
