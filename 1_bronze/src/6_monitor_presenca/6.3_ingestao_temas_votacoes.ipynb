```python
# --------------------------------------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 6.3_ingestao_temas_votacoes
# LOCAL: 1_bronze/src/6_monitor_presenca/
# OBJETIVO: Detecção de padrão de ausência: deputados x faltas em votações específicas. Fonte para ser usada no script 3_gold/../6_monitor_presenca
# ENTREGA: Detecção de padrão de ausência: deputados que faltam mais em votações específicas
# --------------------------------------------------------------------------------------------------------------------------------------------------------

from pyspark.sql import Row

temas_lista = []
# 30 IDs para 3 temas principais
for i in range(999001, 999031):
    if i <= 999010: tema = "Economia"
    elif i <= 999020: tema = "Educação"
    else: tema = "Meio Ambiente"
    
    temas_lista.append(Row(id_votacao=i, tema=tema, ementa=f"Projeto de Lei {i}"))

# 2. Salvar na Bronze
df_temas = spark.createDataFrame(temas_lista)
df_temas.write.mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze_temas_votacoes")

print("Sucesso! Tabela 'bronze_temas_votacoes' criada com temas simulados.")
display(df_temas)
```
