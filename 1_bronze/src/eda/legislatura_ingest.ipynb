```python
# ----------------------------------------------------------------------------------------------------------
# SCRIPT: legislatura_ingest
# LOCAL: 1_bronze/src/eda/
# OBJETIVO: Extração de dados sobre Legistraturas da API da Câmara dos Deputados (usado durante o EDA)
# ----------------------------------------------------------------------------------------------------------

import requests as req

# Acesso a API e recuperação dos dados
response = req.get("https://dadosabertos.camara.leg.br/api/v2/legislaturas").json()
dados_lista = response['dados'] 

# Dataframe para salvar os dados
df_legislatura = spark.createDataFrame(dados_lista)

# Salvando os dados na tabela da camada bronze
df_legislatura.write.mode("overwrite").saveAsTable("bronze_legislatura")
```
