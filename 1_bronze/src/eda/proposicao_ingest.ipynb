```python
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# !!! ---- ESTE SCRIPT NAO SERA MAIS UTILIZADO ---- !!!
# SCRIPT: 3.1a_proposicao_ingest
# LOCAL: 1_bronze/src/3_correlacoes_frentes_votos/
# OBJETIVO: Extração dos dados da Câmara de Deputados sobre proposições legislativas
# ENTREGA: Verificar se deputados de uma mesma frente votam de forma mais alinhada do que seus colegas de partido.
# !!! ---- ESTE SCRIPT NAO SERA MAIS UTILIZADO ---- !!!
# -----------------------------------------------------------------------------------------------------------------------------------------------------

import requests as req

# Acesso a API e recuperação dos dados
response = req.get("https://dadosabertos.camara.leg.br/api/v2/proposicoes?ordem=ASC&ordenarPor=id&itens=1000").json()
dados_lista = response['dados'] 

# Dataframe para salvar os dados
df_proposicoes = spark.createDataFrame(dados_lista)

# Salvando os dados na tabela da camada bronze
df_proposicoes.write.mode("overwrite").saveAsTable("bronze_proposicoes")
```
