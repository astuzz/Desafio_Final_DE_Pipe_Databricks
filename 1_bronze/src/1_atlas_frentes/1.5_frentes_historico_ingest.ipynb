```python
# ------------------------------------------------------------------------------------------------------------------------
# SCRIPT: 1.5_frentes_historico_ingest
# LOCAL: /1_bronze/src/1_atlas_frentes/
# OBJETIVO: Ingestao de frentes legislatura 56 (fonte usada na 3_gold/src/1_atlas_frentes/1.5_evolucao_tematica_frentes)
# ENTREGA: Evolução do número de frentes por tema entre legislaturas
# ------------------------------------------------------------------------------------------------------------------------

import requests as req

# Capturando frentes da legislatura 56 (anterior)
url_56 = "https://dadosabertos.camara.leg.br/api/v2/frentes?idLegislatura=56"
dados_56 = req.get(url_56).json()['dados']

df_frentes_56 = spark.createDataFrame(dados_56)

# Salvando na Bronze 
df_frentes_56.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze_frentes_legislat_56")
```
