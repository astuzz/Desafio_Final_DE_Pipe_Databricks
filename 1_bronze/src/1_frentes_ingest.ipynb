```python
import requests as req

# 1. Coleta e extrai a lista correta
response = req.get("https://dadosabertos.camara.leg.br/api/v2/frentes").json()
dados_lista = response['dados'] 

# 2. Cria o DataFrame (o Spark lida melhor com listas de dicionários assim)
df_frentes = spark.createDataFrame(dados_lista)

# 3. Salva como tabela
df_frentes.write.mode("overwrite").saveAsTable("bronze_frentes")
```
