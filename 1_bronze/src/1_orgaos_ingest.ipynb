```python
import requests as req

# Acesso a API e recuperação dos dados
response = req.get("https://dadosabertos.camara.leg.br/api/v2/orgaos").json()
dados_lista = response['dados'] 

# Dataframe para salvar os dados
df_orgaos = spark.createDataFrame(dados_lista)

# Salvando os dados na tabela da camada bronze
df_orgaos.write.mode("overwrite").saveAsTable("bronze_orgaos")
```
