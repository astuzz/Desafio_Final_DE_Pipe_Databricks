```python
import requests as req

# Acesso a API e recuperação dos dados
response = req.get("https://dadosabertos.camara.leg.br/api/v2/referencias/tiposEvento").json()
dados_lista = response['dados'] 

# Dataframe para salvar os dados
df_referencia_tipoevento = spark.createDataFrame(dados_lista)

# Salvando os dados na tabela da camada bronze
df_referencia_tipoevento.write.mode("overwrite").saveAsTable("bronze_referencia_tipoevento")

display(spark.table("bronze_referencia_tipoevento"))
```
